```python
class Inventario:
    def __init__(self):
        self.inventario = []
        
    def agregar_inventario(self, unidades, costo):
        self.inventario.append({"unidades": unidades, "costo": costo})
        
    def consumir(self, unidades):
        unidades_restantes = unidades
        costo_total = 0
        
        while unidades_restantes > 0 and self.inventario:
            lote = self.inventario[0]
            if lote["unidades"] <= unidades_restantes:
                # Consumir todo el lote
                unidades_restantes -= lote["unidades"]
                costo_total += lote["unidades"] * lote["costo"]
                self.inventario.pop(0)
            else:
                # Consumir parcialmente el lote
                costo_total += unidades_restantes * lote["costo"]
                lote["unidades"] -= unidades_restantes
                unidades_restantes = 0
                
        return costo_total
    
    def valor_total_inventario(self):
        return sum(lote["unidades"] * lote["costo"] for lote in self.inventario)
    
    def mostrar_inventario(self):
        print("\nInventario actual:")
        for lote in self.inventario:
            print(f"Unidades: {lote['unidades']}, Costo unitario: ${lote['costo']}")

# Crear instancia y realizar operaciones
inv = Inventario()

# Inventario inicial
inv.agregar_inventario(350, 85)
print("Después del inventario inicial:")
inv.mostrar_inventario()

# Compra
inv.agregar_inventario(55, 87)
print("\nDespués de la compra:")
inv.mostrar_inventario()

# Consumo
costo_consumo = inv.consumir(370)
print(f"\nCosto del consumo: ${costo_consumo}")
print("Después del consumo:")
inv.mostrar_inventario()

# Devolución
inv.agregar_inventario(50, 87)
print("\nDespués de la devolución:")
inv.mostrar_inventario()

# Valor final del inventario
valor_final = inv.valor_total_inventario()
print(f"\nValor total del inventario final: ${valor_final}")
```
